```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import networkx as nx
```

```python
prefectures = ['Hokkaido', 'Aomori', 'Iwate', 'Miyagi', 'Akita', 'Yamagata',
       'Fukushima', 'Ibaraki', 'Tochigi', 'Gunma', 'Saitama', 'Chiba', 'Tokyo',
       'Kanagawa', 'Niigata', 'Toyama', 'Ishikawa', 'Fukui', 'Yamanashi',
       'Nagano', 'Gifu', 'Shizuoka', 'Aichi', 'Mie', 'Shiga', 'Kyoto', 'Osaka',
       'Hyogo', 'Nara', 'Wakayama', 'Tottori', 'Shimane', 'Okayama',
       'Hiroshima', 'Yamaguchi', 'Tokushima', 'Kagawa', 'Ehime', 'Kochi',
       'Fukuoka', 'Saga', 'Nagasaki', 'Kumamoto', 'Oita', 'Miyazaki',
       'Kagoshima', 'Okinawa']
adj = np.load('../data/adjacency_matrix.npy')
np.fill_diagonal(adj, 1)
```

```python
df = pd.DataFrame(adj, index=prefectures, columns=prefectures)
df
```

```python
fig = plt.figure(figsize=(10, 10))
sns.heatmap(df, linewidths=.3, cbar=False)
plt.savefig('adjcency.png')
```

```python
adj = np.load('../data/adjacency_matrix.npy')
df1 = pd.DataFrame(adj, index=prefectures, columns=prefectures)
G = nx.from_pandas_adjacency(df1)
print(nx.info(G))
```

```python
# pos = nx.kamada_kawai_layout(G)
fig = plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, k=0.45, scale=1.0, seed=500)
# nx.draw_networkx_nodes(G, pos=pos, nodelist=prefectures)
nx.draw_networkx(G, pos=pos, node_size=1200, font_size=10, alpha=0.9)
plt.axis("off")
plt.savefig('networkx.png')
```

```python
# pos = nx.kamada_kawai_layout(G)
fig = plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, k=0.45, scale=1.0, seed=10)
# nx.draw_networkx_nodes(G, pos=pos, nodelist=prefectures)
nx.draw_networkx(G, pos=pos, node_size=1200, font_size=10)
plt.axis("off")
plt.show()
```

```python
fig = plt.figure(figsize=(10, 10))
pos = nx.spring_layout(G, seed=648)  # Seed layout for reproducible node positions
nx.draw(G, pos)
plt.show()
```
