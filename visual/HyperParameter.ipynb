```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
Memory = pd.read_csv('./HyperParameter-Memory.csv')
```

```python
Memory = Memory[['event', 'mem_num', 'mem_dim', 'RMSE', 'MAE']]
Memory.columns=['event', 'Num_Memory', 'Dim_Memory', 'RMSE', 'MAE']
Memory1 = Memory[Memory['Dim_Memory']==8]
Memory1 = Memory1.sort_values(by='Num_Memory')
Memory1['Num_Memory'] = Memory1['Num_Memory'].astype(str)
Memory2 = Memory[Memory['Num_Memory']==10]
Memory2 = Memory2.sort_values(by='Dim_Memory')
Memory2['Dim_Memory'] = Memory2['Dim_Memory'].astype(str)
```

```python
sns.set_style("darkgrid")
fig, ax =plt.subplots(1,2,figsize=(8,3.5))
sns.lineplot(data=Memory1, x='Num_Memory', y='RMSE', hue='event', style='event', markers=True, dashes=False, ax=ax[0], linewidth=2, markersize=10)
sns.lineplot(data=Memory2, x='Dim_Memory', y='RMSE', hue='event', style='event', markers=True, dashes=False, ax=ax[1], linewidth=2, markersize=10)
ax[0].set_title('Average Prediction Error', fontsize=12)
ax[0].legend(title=None, fontsize=12)
ax[0].xaxis.set_tick_params(labelsize=12)
ax[0].yaxis.set_tick_params(labelsize=12)
ax[0].xaxis.label.set_size(12)
ax[0].yaxis.label.set_size(12)
# ax[0].set(ylim=(0, None))
# ax[0].set_xticks(Memory["Num_Memory"].unique())
ax[1].set_title('Average Prediction Error', fontsize=12)
ax[1].legend(title=None, fontsize=12)
ax[1].set(ylabel=None)
ax[1].xaxis.set_tick_params(labelsize=12)
ax[1].yaxis.set_tick_params(labelsize=12)
ax[1].xaxis.label.set_size(12)
ax[1].yaxis.label.set_size(12)
plt.tight_layout(pad=0.4)
# fig.subplots_adjust(wspace=0.25)
plt.savefig('HyperParameter-Memory.png')
```

```python
Step = pd.read_csv('./HyperParameter-Step.csv')
Step = Step[['event', 'timestep', 'flag', 'flag1', 'RMSE', 'MAE']]
Step = Step.sort_values(by='timestep')
Step['timestep'] = Step['timestep'].astype(str)
Step.columns=['event', 'alpha/beta (timestep)', 'flag', 'flag1', 'RMSE', 'MAE']
Avg = Step[Step.flag1=='average']
OneHour = Step[Step.flag1=='1 hour ahead']
ThreeHour = Step[Step.flag1=='3 hours ahead']
print(Avg.shape, OneHour.shape, ThreeHour.shape)
```

```python
fig, ax =plt.subplots(1,3,figsize=(10,4))
sns.lineplot(data=Avg, x='alpha/beta (timestep)', y='RMSE', hue='event', style='event', markers=True, dashes=False, ax=ax[0], linewidth=2, markersize=10)
sns.lineplot(data=OneHour, x='alpha/beta (timestep)', y='RMSE', hue='event', style='event', markers=True, dashes=False, ax=ax[1], linewidth=2, markersize=10)
sns.lineplot(data=ThreeHour, x='alpha/beta (timestep)', y='RMSE', hue='event', style='event', markers=True, dashes=False, ax=ax[2], linewidth=2, markersize=10)
ax[0].legend(title=None, fontsize=12)
ax[0].xaxis.set_tick_params(labelsize=12)
ax[0].yaxis.set_tick_params(labelsize=12)
ax[0].xaxis.label.set_size(12)
ax[0].yaxis.label.set_size(12)
ax[0].set_title('Average Prediction Error', fontsize=12)
# ax[0].set(ylim=(0, None))
# ax[0].set_xticks(Memory["Num_Memory"].unique())

ax[1].legend(title=None, fontsize=12)
ax[1].set(ylabel=None)
ax[1].xaxis.set_tick_params(labelsize=12)
ax[1].yaxis.set_tick_params(labelsize=12)
ax[1].xaxis.label.set_size(12)
ax[1].yaxis.label.set_size(12)
ax[1].set_title('1 Hour Ahead Prediction Error', fontsize=12)

ax[2].legend(title=None, fontsize=12)
ax[2].set(ylabel=None)
ax[2].xaxis.set_tick_params(labelsize=12)
ax[2].yaxis.set_tick_params(labelsize=12)
ax[2].xaxis.label.set_size(12)
ax[2].yaxis.label.set_size(12)
ax[2].set_title('3 Hours Ahead Prediction Error', fontsize=12)

plt.tight_layout(pad=0.4)
plt.savefig('HyperParameter-Step.png')
```
